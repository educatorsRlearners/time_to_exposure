## Filter for children who have done at least 4 words at least 2 lexile levels

```python
pivoted.groupby("childid")['lexile'].nunique().reset_index().sort_values(by='lexile', ascending=False)
```

```python
child_performance = pivoted.groupby(['childid', 'lexile']).aggregate({'reading_age': 'count', 'err_4':'sum'}).reset_index().rename(columns={'reading_age':'word_count', 'err_4':'fail_count'})
child_performance['success_count'] = child_performance['word_count'] - child_performance['fail_count']
child_performance['success_rate'] = child_performance['success_count'] / child_performance['word_count']
child_performance['fail_rate'] = child_performance['fail_count'] / child_performance['word_count']
```

```python
child_performance.head(15)
```

## Keep Kids with a Minimum of 4 words at a Lexile Level

```python
four_words = child_performance[child_performance['word_count'] >= 4]
four_words.shape
```

## Keep Kids with a Minimum of Two Lexile Levels

```python
# Find duplicated childid values
duplicated_childid = four_words['childid'].duplicated(keep=False)

# Filter the DataFrame based on duplicated childid
four_words_2x = four_words[duplicated_childid]

# Display the counts to verify results
four_words_2x.groupby('childid')['childid'].count().sort_values(ascending=False)
```

## Identify Children's Rate Change

```python
four_words_2x.loc[:, 'rate_change'] = four_words_2x.groupby('childid')['fail_rate'].diff() > 0

#Replace the first rate_change for each child with "N/A"
four_words_2x.loc[four_words_2x.groupby('childid').head(1).index, 'rate_change'] = "N/A"

four_words_2x.head(25)
```

```python
# Determine if rate_change is consistently increasing for each child. 
# That would indicate the child is performing worse on the harder words. 
consistent_increase = four_words_2x.groupby('childid')['rate_change'].all()

# Count the children with consistently increasing rate_change
count_increasing = consistent_increase[consistent_increase].count()

print("Count of children with consistently increasing fail rate:", count_increasing)
```

```python
consistent_increase.value_counts()
```

```python
df_results = pd.merge(left=four_words_2x,
                      right=consistent_increase.reset_index().rename(columns={'rate_change': 'monotonic'}),
                      how='inner',
                      on='childid')
df_results.head(20)
```

## 1/3 of Children's observations are Monotonicly increasing in Fail Rate By Lexile

```python
df_results['min_lexile'] = df_results.groupby(['childid'])['lexile'].transform('min')
df_results['relative_lexile'] = df_results['lexile'] - df_results['min_lexile']

df_results.head(10)
```
